```python
import pandas as pd
```

Lendo o arquivo csv

```python
basis_risk_credit  = pd.read_csv("/content/riscoCredito.csv",encoding = "UTF-8", sep = ";")
```

```python
basis_risk_credit 
```

Fazendo o processamentos dos dados, 
fazendo a separação de X e Y


```python
X_risk_credit = basis_risk_credit.iloc[:, 0:4].values
X_risk_credit
```

```python
Y_risk_credit = basis_risk_credit.iloc[:,4].values
Y_risk_credit
```

Transformando strings em números para aplica os métodos.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder_casapropria = LabelEncoder()
label_encoder_estadocivil = LabelEncoder()
label_encoder_possuicarro = LabelEncoder()
label_encoder_rendimentos = LabelEncoder()

```

Aplicando os processamentos

```python
X_risk_credit[:,0] = label_encoder_casapropria.fit_transform(X_risk_credit[:,0])
X_risk_credit[:,1] = label_encoder_estadocivil.fit_transform(X_risk_credit[:,1])
X_risk_credit[:,2] = label_encoder_possuicarro.fit_transform(X_risk_credit[:,2])
X_risk_credit[:,3] = label_encoder_rendimentos.fit_transform(X_risk_credit[:,3])
```

```python
X_risk_credit
```

Gravando os dados formatados, Anteriormente 

```python
import pickle
with open('risk_credit.pkl', 'wb') as f:
  pickle.dump([X_risk_credit,Y_risk_credit],  f)
```

Usando o método niave-Bayes

```python

from sklearn.naive_bayes import GaussianNB
```

Treinando o algoritomo

```python
naive_risk_credit = GaussianNB()
naive_risk_credit.fit(X_risk_credit,Y_risk_credit)
```

Queremos saber se um cliente que não possui casa própria [1], cujo estado civil é divorciado [1], possui carro [1] e tem um rendimento alto [0], é um bom pagador:

```python
forecast = naive_risk_credit.predict([[1, 1, 1, 0]])
print(forecast)
```

Queremos saber se um cliente que possui casa própria [1], cujo estado civil é Solteiro [2], possui carro [0] e tem um rendimento Medio [2], é um bom pagador:

```python
forecast = naive_risk_credit.predict([[1, 2, 0, 2]])
print(forecast)
```

O cliente de registro 12 não seria um bom pagador (Sim) e o cliente de registro 13 seria um bom pagador(SIM)
